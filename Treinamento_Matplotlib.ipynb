## Introdução
O Matplotlib é baseado em Figuras (_figures_), que podem conter um ou mais eixos (onde serão, de fato, plotados os gráficos).

A construção de um gráfico se baseia em um estado interno que pode ser construído passo a passo. Ao fim, é possível salvar a figura ou exibi-la.

No primeiro exemplo, será criada uma figura simples, com um único eixo. Depois, será demonstrado como trabalhar com múltiplos gráficos (eixos) na mesma figura.

Obs.: No caso, o eixo referido é o eixo de plotagem, e não (X, Y).

### Figure
Para iniciar um gráfico, é interessante a criação do objeto figura, através do seguinte comando.
    
    plt.figure()

Parâmetros de interesse:
- Figsize: figsize = (largura, altura)

Após a criação da figura, seguiremos com a criação dos gráficos.

### Plot
Função: Plota os valores y versus x em um gráfico de linhas podendo conter marcadores.

Utilização básica: 

    plt.plot([X], Y) # onde o X é opcional.

> Referência: [Matplotlib plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot)

A função _plot_ pode ser chamada diretamente (plt.plot()) ou especificando sobre qual eixo irá atuar (ax1.plot(), ax2.plot()).

Estas 2 situações serão demonstradas abaixo.

#### Parâmetros (plot)

Além da plotagem básica, pode-se inserir parâmetros opcionais para customizar os gráficos. Alguns destes serão demonstrados a seguir.

Exemplos:
- linewidth: largura da linha
- linestyle: estilo da linha
- color: cor da linha
- marker: tipo de marcador
- markersize: tamanho de marcador
- markerfacecolor: cor do marcador

### Show
Exibe a(s) figura(s) criada(s).

    plt.show()

### Savefig
Utiliza-se, ao fim da construção da figura, para salvá-la.

    plt.savefig(fname)

Parâmetros obrigatórios:
- fname (string): pode ser apenas o nome da imagem, salvando no diretório atual, ou o caminho completo de onde a imagem deve ser armazenada.

Parâmetros opcionais:
- dpi (float): resolução da imagem em unidades de pontos por polegada. 
- format (string): formato da imagem.

### Subplots 
[Referência - plt.subplots()](https://matplotlib.org/devdocs/api/_as_gen/matplotlib.pyplot.subplots.html#matplotlib.pyplot.subplots)

Através de subplots é possível plotar diversos gráficos em uma figura única.

Comando básico: 

    plt.subplots(n_linhas, n_colunas)
O comando acima retorna uma figura e um conjunto de subplots (eixos).\
Iremos, então, atribuir variáveis a esses elementos para que possamos trabalhar com os mesmos, através dos seguintes comandos.

    fig, axs = plt.subplots(n_linhas, n_colunas) # Cria uma figura (fig) com n eixos acessados através de axs[n_linha, n_coluna]

    fig, (ax1, ax2) = plt.subplots(2, 1) # Cria uma figura (fig) com 2 eixos (ax1, ax2)

Para acessar um dos n eixos, utiliza-se o seguinte comando:

    axs[n_linha, n_coluna]

Onde [n_linha, n_coluna] é referente a posição dos gráficos dentro da imagem. Ambos iniciam a contagem em zero.

Os seguintes parâmetros podem ser de interesse.
- figsize(float, float) - indica o tamanho da figura a ser criada (default: [6.4, 4.8]).
- sharex, sharey
    - booleano - indica se os eixos cartesianos X ou Y serão compartilhados pelos subplots.
    - raw/col (string) - Os eixos cartesianos passar a ser compartilhados nas linhas ou colunas.

> Exemplo: **sharey = 'raw'** indica que o eixo Y será o mesmo para os gráficos pertencentes a uma mesma linha.

Pode-se observar a utilização básica dessa funcionalidade no exemplo a seguir.

Outras aplicações/customizações de subplots podem ser verificadas através da referência indicada acima.


```python
# Figura com 2 eixos (gráfico)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,8))  # Cria uma figura (fig) com 2 eixos (ax1, ax2)

ax1.plot([2, 4, 8, 16]) # Plota os valores em ax1
ax2.plot([1, 32, 3, 10]) # Plota os valores em ax1

plt.show() # Exibe a figura
```

Será utilizado o seguinte comando para criar uma figura com 6 gráficos dispostos em 3 linhas e 2 colunas.

    fig, axs = plt.subplots(3, 2, figsize = (8,10))

Para acessar um dos 6 eixos, utiliza-se o seguinte comando:

    axs[n1, n2]

Onde n1 e n2 são referentes a posição dos gráficos dentro da imagem. n1 é referente a linha e n2, coluna. Ambos iniciam a contagem em zero.

Portanto, para acessar o gráfico na terceira linha e segunda coluna, utilizariamos os parâmetros n1 = 2 e n2 = 1.

```python
y = [2, 4, 6, 8]

fig, axs = plt.subplots(3, 2, figsize = (8,10))  # Cria uma figura (fig) com 6 eixos (ax1, ax2) e tamanho de (8, 10)

axs[0, 0].plot(y) # Plota os valores no gráfico na primeira linha/coluna
axs[0, 0].set_title("Plotagem padrão") # Define o título do gráfico 

axs[1, 0].plot(y, linewidth = '5') # Plota os valores no gráfico da segunda linha e primeira coluna
axs[1, 0].set_title("linewidth = '5'") 

axs[2, 0].plot(y, linestyle = '--') # Plota os valores no gráfico da terceira linha e primeira coluna
axs[2, 0].set_title("linestyle = '--'")

axs[0, 1].plot(y, marker = 'o') # Plota os valores no gráfico da primeira linha e segunda coluna
axs[0, 1].set_title("marker = 'o'")

axs[1, 1].plot(y, marker = 'D', markersize='14') # Plota os valores no gráfico da segunda linha e segunda coluna
axs[1, 1].set_title("marker = 'D', markersize='14'")

axs[2, 1].plot(y, marker = 'o', markersize='10', markerfacecolor='red') # Plota os valores no gráfico da terceira linha e segunda coluna
axs[2, 1].set_title("marker = 'o', markersize='10',\nmarkerfacecolor='red'")

# Define o título global da figura, assim como desloca-o para cima com y=1 (default y = 0.95).
fig.suptitle("plt.plot: Parâmetros Adicionais", y=1) 

# Ajusta os espaçamentos e afins, para não haver sobreposição de informações
plt.tight_layout()

# Exibe o(s) gráfico(s) plotado(s)
plt.show()
```

### Gráfico de dispersão
[Referência: Scatter Plots Matplotlib](https://matplotlib.org/stable/gallery/lines_bars_and_markers/scatter_with_legend.html#sphx-glr-gallery-lines-bars-and-markers-scatter-with-legend-py)

Neste tipo de gráfico pode-se verificar a relação entre 2 variáveis quantitativas. 

Através deste gráfico, podem ser obtidas valiosas informações ao observar a relação entre variáveis, como a formação de grupos ou detecção de _outliers_.

```python
# Dados
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

# Plotagem
fig, ax = plt.subplots(figsize = (12,5))
ax.scatter(x, y)

# Definindo o título e as legendas dos eixos (labels)
plt.suptitle("Gráfico de Dispersão Simples", fontsize=14)
plt.xlabel('Variável 1', fontsize=14)
plt.ylabel('Variável 2', fontsize=14)

plt.tight_layout()
plt.show()
```

No exemplo abaixo, há a inserção de mais 2 parâmetros que podem ser definidos de acordo com mais variáveis de interesse. São eles: cor e tamanho do "ponto".

```python
# Gerando dados aleatórios (dados de X, Y, além de cores e tamanho)
N = 45
x, y = np.random.rand(2, N)
cores = np.random.randint(1, 5, size=N)
tamanhos = np.random.randint(10, 220, size=N)

# Criando a figura e plotando os dados
fig, ax = plt.subplots(figsize=(10,5))
scatter = ax.scatter(x, y, c=cores, s=tamanhos)

# Criando legenda para as cores
legend1 = ax.legend(*scatter.legend_elements(), loc="lower left", title="Classes")
ax.add_artist(legend1)

# Criando legenda para os tamanhos
handles, labels = scatter.legend_elements(prop="sizes", alpha=0.6)
legend2 = ax.legend(handles, labels, loc="upper right", title="Sizes")

plt.suptitle("Gráfico de Dispersão com Legendas")
plt.show()
```

### Gráfico de Barras
Este tipo de gráfico é bastante utilizado visualização de dados onde há classes definidas e representadas por variáveis não-contínuas. Assim, pode-se comparar as classes em relação ao nível de ocorrência das mesmas.

```python
grupos = ['G1', 'G2', 'G3', 'G4', 'G5']
pontuacao_homens = [20, 34, 30, 35, 27]
pontuacao_mulheres = [25, 32, 34, 20, 25]

x = np.arange(len(grupos))  # Localização dos nomes dos grupos
width = 0.35  # Largura das barras

fig, ax = plt.subplots(figsize=(10,6))
rects1 = ax.bar(x - width/2, pontuacao_homens, width, label='pontuacao_homens')
rects2 = ax.bar(x + width/2, pontuacao_mulheres, width, label='pontuacao_mulheres')

# Definindo título da figura, nomes dos eixos, marcadores do eixo X e habilitando a legenda
ax.set_ylabel('Pontuação')
ax.set_xlabel('Grupo')
ax.set_title('Pontuação por grupo e por gênero')
ax.set_xticks(x, grupos)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

plt.tight_layout()
plt.show()
```

### Histogramas
Os histogramas caracterizam-se por dispor uma **variável contínua** com base em sua **distribuição de frequência**.

O gráfico de barras, por sua vez, representa variáveis discretas/categóricas.

Os histogramas podem fornecer informações sobre:
- Distribuição dos valores de um conjunto de dados
- Frequência de ocorrência dos dados
- Valores mais frequentes
- Tendência central
- Amplitude
- Simetria

O número de classes pode impactar a análise sobre os dados em questão.

Exemplo: Distribuição das idades de pessoas que compareceram em determinado evento.
- Classe 1 - número de pessoas entre 10 e 20 anos
- Classe 2 - número de pessoas entre 20 e 30 anos
- Classe 3 - número de pessoas entre 30 e 40 anos.

```python
# Gerando os dados
idades = np.array([22,5,43,56,73,55,54,11,20,51,5,79,31,27,1,5,90,7,10,13,18])
idades_limite = [0,20,40,60,80,100] # Estas são as idades que irão delimitar os 'grupos'

# Criando a figura e plotando os dados
fig,ax = plt.subplots(1,1, figsize=(10,5))
ax.hist(x = idades, bins = idades_limite)

# Customizando o gráfico
ax.set_title("Distribuição de Idades em determinado evento", fontsize=14)
ax.set_xticks(idades_limite)
ax.set_xlabel('Idade', fontsize=14)
ax.set_ylabel('Quantidade de pessoas', fontsize=14)

plt.grid()
plt.show()
```

```python
caminho = 'https://raw.githubusercontent.com/joceliovieira/treinamento_data_viz/main/data/rad_jp.csv'
df_jp = pd.read_csv(caminho,
                    delimiter=';',
                    names=['DATA','HORA','RAD_KJM2'],
                    header=9, 
                    decimal=',',
                    index_col= False
                    )
df_jp.head()
```

### Análise exploratória e tratamento inicial dos dados

```python
df_jp.info()
```

```python
df_jp.head(10)
```

```python
# Padronizando o formato de dados, para converter para hora e depois alterar o fuso horário
list = []
for i in df_jp['HORA']:
    if i == 0:
        a = '000' + str(i)
        list.append(a)
    elif i in [100,200,300,400,500,600,700,800,900]: 
        a = '0' + str(i)
        list.append(a)    
    else:
        a = str(i)
        list.append(a)

print(list[:24])
```

### Atribuindo uma Coluna para os dados de hora 'convertidos' e criando uma coluna para data+hora, que servirá de índice (index)

```python
df_jp['HORA_CONV'] = list

df_jp['DATA_HORA'] = df_jp['DATA'] + ' ' + df_jp['HORA_CONV']
df_jp.head()
```

Removendo as colunas não mais necessárias (data, hora e hora_conv), e convertendo o formato de data_hora.

```python
df_jp.drop(columns=['DATA', 'HORA', 'HORA_CONV'], inplace=True)
df_jp.head()
```

Convertendo o tipo de data_hora de String para Datetime

```python
# Convertendo de strings para date_time
print(df_jp.info())
df_jp['DATA_HORA'] = pd.to_datetime(df_jp['DATA_HORA'],format="%Y-%m-%d %H%M")
print(df_jp.info())
```

#### Convertendo o fuso horário e atribuindo a coluna data_hora ao índice do Dataframe


```python
df_jp.set_index(df_jp['DATA_HORA'], inplace=True)
df_jp['DATA_HORA_RECIFE'] = df_jp['DATA_HORA'].apply(lambda x: df_jp['DATA_HORA'][x].tz_localize('UTC').tz_convert('America/Recife'))
df_jp = df_jp.set_index(df_jp['DATA_HORA_RECIFE']).drop(columns=['DATA_HORA','DATA_HORA_RECIFE'])
df_jp.index.names = ['DATA_HORA']
df_jp.head()
```

Removendo os itens iniciais anteriores a 2010


```python
df_jp.drop(df_jp['RAD_KJM2'].index[:3], inplace=True)
df_jp.head()
```

Convertendo as unidades de radiação e criando uma coluna para os dados em kWh/m²

```python
df_jp['RAD_KWHM2'] = (df_jp['RAD_KJM2']/3600).round(2)
df_jp.drop(columns='RAD_KJM2', inplace = True)
df_jp.head()
```

A técnica acima não apresentou comportamento ideal, uma vez que apenas preencheu com um valor médio grandes intervalos de tempo. Para melhorar tal preenchimento de valores ausentes, pode-se utilizar de técnicas como a interpolação por inverso da distância ou inverse distance weighting (IDW), que relaciona dados de localizações próximas ao dado ausente e fornece um valor aproximado para o mesmo.
